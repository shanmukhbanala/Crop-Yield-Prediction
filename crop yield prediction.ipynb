```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler

df = "/content/drive/MyDrive/crop_yield.csv"
df = pd.read_csv(df)
df.head()

```

```python
##code without interface

import pandas as pd
from sklearn.linear_model import LinearRegression

# Assuming your dataset is stored in a DataFrame called 'df'
# You can load your dataset using pd.read_csv or another appropriate method

# Select features and target variable
X = df[['Crop_Year', 'Area', 'Production', 'Annual_Rainfall', 'Fertilizer', 'Pesticide', 'Crop', 'Season', 'State']]
y = df['Yield']

# Convert categorical variables to numerical using one-hot encoding
X_encoded = pd.get_dummies(X, columns=['Crop', 'Season', 'State'])

# Create a linear regression model
model = LinearRegression()

# Fit the model to the entire dataset
model.fit(X_encoded, y)

# Get user input for new values
crop_year = int(input("Enter Crop Year: "))
area = float(input("Enter Area: "))
production = float(input("Enter Production: "))
annual_rainfall = float(input("Enter Annual Rainfall: "))
fertilizer = float(input("Enter Fertilizer: "))
pesticide = float(input("Enter Pesticide: "))
crop = input("Enter Crop: ")
season = input("Enter Season: ")
state = input("Enter State: ")

# Create a DataFrame for the user input
user_data = pd.DataFrame({
    'Crop_Year': [crop_year],
    'Area': [area],
    'Production': [production],
    'Annual_Rainfall': [annual_rainfall],
    'Fertilizer': [fertilizer],
    'Pesticide': [pesticide],
    'Crop': [crop],
    'Season': [season],
    'State': [state]
})

# Convert categorical variables to numerical using one-hot encoding
user_data_encoded = pd.get_dummies(user_data, columns=['Crop', 'Season', 'State'])

# Ensure the user input DataFrame has the same columns as the training DataFrame
user_data_encoded = user_data_encoded.reindex(columns=X_encoded.columns, fill_value=0)

# Make predictions for the user input
predicted_yield = model.predict(user_data_encoded)

print(f'Predicted Yield: {predicted_yield[0]}')
```

```python


from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


y_pred = model.predict(X_encoded)

mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R2): {r2}")
```

```python
#for interface

import gradio as gr

def predict_yield(crop_year, area, production, annual_rainfall, fertilizer, pesticide, crop, season, state):
    # Create a DataFrame for the user input
    user_data = pd.DataFrame({
        'Crop_Year': [crop_year],
        'Area': [area],
        'Production': [production],
        'Annual_Rainfall': [annual_rainfall],
        'Fertilizer': [fertilizer],
        'Pesticide': [pesticide],
        'Crop': [crop],
        'Season': [season],
        'State': [state]
    })

    # Convert categorical variables to numerical using one-hot encoding
    user_data_encoded = pd.get_dummies(user_data, columns=['Crop', 'Season', 'State'])

    # Ensure the user input DataFrame has the same columns as the training DataFrame
    user_data_encoded = user_data_encoded.reindex(columns=X_encoded.columns, fill_value=0)

    # Make predictions for the user input
    predicted_yield = model.predict(user_data_encoded)

    return f'Predicted Yield: {predicted_yield[0]}'

# Create Gradio interface
iface = gr.Interface(fn=predict_yield,
                     inputs=['text', 'number', 'number', 'number', 'number', 'number', 'text', 'text', 'text'],
                     outputs='text')

# Launch the interface
iface.launch()
```
